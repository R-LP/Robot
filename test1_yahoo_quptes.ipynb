```python
import pandas_datareader as pdr
import pandas as pd
import datetime 
cgi = pdr.get_data_yahoo('CGI', 
                          start=datetime.datetime(2017, 1, 1), 
                          end=datetime.datetime(2017, 9, 16))


cgi.to_csv('cgi_ohlc.csv')
```

```python
# Sample 20 rows
sample = cgi.sample(20)

# Print `sample`
print(sample)

# Resample to monthly level 
monthly_cgi = cgi.resample('W').mean()

cgi.resample?

# Print `monthly_aapl`
print(monthly_cgi)
```

```python
import matplotlib.pyplot as plt

cgi['Close'].plot(grid=True)

# Show the plot
plt.show()
cgi.index
```

```python
open=[]
close=[]
high=[]
low=[]

for i in cgi.Open:
    open.append(i)
    
for i in cgi.High:
    high.append(i)
    
for i in cgi.Low:
    low.append(i)  

for i in cgi['Adj Close']:
     close.append(i)
```

```python
import matplotlib.dates
date=[]

for i in cgi.index:
    date.append(matplotlib.dates.date2num(i))    
```

```python
quotes = [tuple([date[i],
                 open[i],
                 high[i],
                 low[i],
                 close[i]]) for i in range(len(date))]
```

```python
import matplotlib.ticker as mticker
from matplotlib.finance import candlestick_ohlc
import matplotlib.dates as mdates



fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')

f1 = plt.subplot2grid((6, 4), (1, 0), rowspan=6, colspan=4)
candlestick_ohlc(f1, quotes, width=0.6, colorup='#53c156', colordown='#ff1717')

f1.xaxis.set_major_formatter(mdates.DateFormatter('%y-%m-%d'))
f1.grid(True)

plt.xticks(rotation=45)
plt.ylabel('Stock Price')
plt.xlabel('Date')
plt.savefig('Test.png')
plt.show()
```
